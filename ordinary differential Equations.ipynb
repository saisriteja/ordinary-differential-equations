# 1.Exponential decay and growth are one of the major important application of ODE

```python
from math import *
#derivative fuction
def derive(function, value):
    h = 0.00000000001
    top = function(value + h) - function(value)
    bottom = h
    slope = top / bottom
    # Returns the slope to the third decimal
    return float("%.3f" % slope)
#exponential function
def f1(x):
    return  exp(x)
def f2(x):
    return exp(-1*x)

   
dresult1=[]
dresult2=[]
dresult3=[]
dresult4=[]
#to plot graph for (0-20)
for i in range(-9000, 9000, 1):
    z1=derive(f1,i/1000)
    z2=derive(f2,i/1000)
    z3=f1(i/1000)
    z4=f2(i/1000)
    
    dresult1.append(z1)
    dresult2.append(z2)
    dresult3.append(z3)
    dresult4.append(z4)
a=np.arange(-9,9,0.001)

plt.subplot(2, 2, 1)
plt.plot(a,dresult1,'r')
plt.title("diff e^(x)")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(a,dresult2,'g')
plt.title("diff e^(-x)")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(a,dresult3,'b')
plt.title(" e^(x)")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(a,dresult4,'p')
plt.title('e^(-x)')
plt.grid(True)
#helps to fit all subplots in the output shown.
plt.tight_layout()

plt.show()
```

```python
#run this in local python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro', animated=True)

def init():
    ax.set_xlim(0, 1000000000000000000)
    ax.set_ylim(-1, 1000000000000000000)
    return ln,

def update(frame):
    print("frmae",'\n',frame)
    z=int(np.exp(frame))
    print("z",'\n',z)
    for x in range(z):
        x1=random.randint(1,1000000000000000000)
        y1=random.randint(1,1000000000000000000)
        xdata.append(x1)
        ydata.append(y1)
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 9, 128),
                    init_func=init, blit=True)

plt.show()
```

```python
#run this in local python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from random import randint

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro', animated=True)

def init():
    ax.set_xlim(0, 1000000000000000000)
    ax.set_ylim(-1, 1000000000000000000)
    return ln,

def update(frame):
    print("frmae",'\n',frame)
    z=int(np.exp(-1*frame))
    print("z",'\n',z)
    
    x1=[randint(1,1000000000000000000)  for x in range(z) ]
    y1=[randint(1,1000000000000000000)  for x in range(z)]
    xdata=x1
    ydata=y1
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(-9, 0, 50),
                    init_func=init, blit=True)
plt.show()
```
